```python
#import packages
import pandas as pd 
import numpy as np
import plotly.express as px
import plotly
```

```python
#load datasets
df1  = pd.read_csv('Isocodes1.csv')

df2 = pd.read_csv('MensNationalTeamRankings - Sheet1.csv')

df3 = pd.read_csv('WomensNationalTeamRankings - Sheet1.csv')
```

```python
#merge Women's and Men's FIFA Datasets using outer method to check any matching errors
df = pd.merge(df2,df3, on='Country Name', how = 'outer')
```

```python
#check for any spelling errors
print(df.to_string())
```

```python
#remove countries which do not have a ranked Women's team
df = df.dropna()
```

```python
#merge country code dataset with football dataset using outer method to check for errors
dfcheck = pd.merge(df, df1, on='Country Name', how = 'outer')
```

```python
#check for matching errors
print(dfcheck.to_string())
```

```python
#reset index
dfcheck.reset_index()
#correct any matching errors in excel
#dfcheck.to_csv('IsocodesMatch.csv')
```

```python
#matched dataset
dfmatched = pd.read_csv('Isocodes2.csv')
```

```python
#use outer method to check for matching errors
df = pd.merge(df,dfmatched, how = 'outer')
```

```python
df = df[['Country Name', 'Mens Points', 'Womens Points', 'Alpha-3 code']]
```

```python
#check for matching errors
print(df.to_string())
```

```python
#remove countries which do not have a ranked Men's or Women's Team
df = df.dropna()
```

```python
#create new column calculating the percentage difference between a countrys Mens and Womens FIFA Ranking points 
df['Percentage Difference'] = ((df['Mens Points']-df['Womens Points'])/df['Mens Points'])*100
```

```python
#check dataset
df.head()
```

```python
#save dataset
df.to_csv('IsocodesMergedClean.csv', index= False)
```

```python
#load geojson file
geojson = 'https://raw.githubusercontent.com/python-visualization/folium/main/examples/data/world-countries.json'
```

```python
#plot choropleth map
fig = px.choropleth(df, locations='Alpha-3 code', geojson=geojson, hover_data=['Country Name'], color= 'Percentage Difference',range_color = (-67.5, 45),color_continuous_scale=px.colors.sequential.Jet, title = "Percentage Difference of Country's Men's FIFA Ranking Points compared to their Women's")
#view choropleth map
fig.write_html('first_choro.html', auto_open=True)
```

```python
#download choropleth map as javascript
plotly.offline.plot(fig, include_plotlyjs=False, output_type='div')
```
