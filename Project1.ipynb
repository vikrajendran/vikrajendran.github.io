```python
#import packages
import pandas as pd
import numpy as np 
```

```python
#load datasets
df1 = pd.read_csv('HDIData1 - Sheet1.csv')

df2 = pd.read_csv('IsocodesMergedClean.csv')
```

```python
#rename column
df1.columns = ['Country Name' if x=='Country' else x for x in df1.columns]
```

```python
#merge datasets using outer method to check for matching errors
dfcheck = pd.merge(df1, df2, on ='Country Name', how = 'outer')
```

```python
#check for matching errors
print(dfcheck.to_string())
```

```python
#correct matching errors in excel
dfcheck.to_csv('HDICheck.csv')
```

```python
#load cleaned data
df3 = pd.read_csv('HDIDataCleaned.csv')
```

```python
#rename columns
df3.columns = ['Country Name' if x=='Country' else x for x in df1.columns]
```

```python
#merge using outer to check for any missed errors
df = pd.merge(df2, df3, on ='Country Name', how = 'outer')
```

```python
#check for erros
print(df.to_string())
```

```python
#remove any country without data
df = df.dropna()
```

```python
#download data as csv
df.to_csv('HDIPercentDiff.csv')
```

```python
#format data for regression
X = df['Percentage Difference'].values[:, np.newaxis]

y = df['Value'].values
```

```python
#perform regression
import statsmodels.api as sm


x2 = sm.add_constant(X)

model = sm.OLS(y, x2).fit()
predictions = model.predict(x2)



model.summary()
```
