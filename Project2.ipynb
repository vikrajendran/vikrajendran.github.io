```python
#import packages
import pandas as pd
import numpy as np
```

```python
#load datasets
df1 = pd.read_csv('GenderInequalityData - Sheet1.csv')

df2 = pd.read_csv('IsocodesMergedClean.csv')
```

```python
#merge using outer method to check for matching errors
dfcheck = pd.merge(df1, df2, on = 'Country Name', how = 'outer')
```

```python
#check for matching errors
print(dfcheck.to_string())
```

```python
#download csv to correct errors in excel
dfcheck.to_csv('GenderInequalityMatch.csv')
```

```python
#load cleaned dataset
df3 = pd.read_csv('GenderInequalityDataCleaned.csv')
```

```python
#merge again
df = pd.merge(df2, df3, on = 'Country Name', how = 'outer')
```

```python
#double check for any remaining errors
print(df.to_string())
```

```python
#remove Countries without data
df = df.dropna()
```

```python
#download csv
df.to_csv('GenderPercentDiff.csv', index = False)
```

```python
#format data for regression
X = df['Percentage Difference'].values[:, np.newaxis]

y = df['Gender Inequality Index'].values
```

```python
#perform regression
import statsmodels.api as sm

x2 = sm.add_constant(X)

model = sm.OLS(y, x2).fit()
predictions = model.predict(x2)




model.summary()
```
