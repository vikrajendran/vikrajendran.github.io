```python
#import packages
import pandas as pd
import numpy as np
```

```python
#load datasets
df1 = pd.read_csv('GenderPercentD.csv')

df2 = pd.read_csv('PopulationData1.csv')
```

```python
#merge datasets using outer method to check for matching errors
df = pd.merge(df1,df2, on = 'Alpha-3 code', how = 'outer')
```

```python
#check for matching errors
print(df.to_string())
```

```python
#rename column
df.columns = ['Population' if x=='YR2021' else x for x in df.columns]
```

```python
#remove countries without data
df = df.dropna()
```

```python
#check shape
df.shape
```

```python
#download csv
df.to_csv('GenderPoints.csv', index = False)
```

```python
#format data for regression
X = df['Womens Points'].values[:, np.newaxis]

y = df['Gender Inequality Index'].values
```

```python
#perform regression
import statsmodels.api as sm

x2 = sm.add_constant(X)

model = sm.OLS(y, x2).fit()
predictions = model.predict(x2)




model.summary()
```
